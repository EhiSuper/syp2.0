```python
df = pd.read_csv('smaller_dataset.csv', on_bad_lines='skip')
df.columns = df.columns.str.replace('"', '')
df.columns = df.columns.str.replace('name', '')
df.columns = df.columns.str.replace(' ', '')
df.columns
df.head()
```

```python
df = df.dropna()
df = df.reset_index(drop = True)
df.shape
```

```python
df['artist'] = df['artist'].str.replace('[^\w\s]','', regex = True)
df['track'] = df['track'].str.replace('[^\w\s]','', regex = True)
df['playlist'] = df['playlist'].str.replace('[^\w\s]','', regex = True)
```

```python
df.head()
```

```python
new_df = df.drop(['user_id'], axis=1)
new_df.head()
```

```python
df['artist'] = df['artist'].str.replace("  ", " ")
df['track'] = df['track'].str.replace("  ", " ")
df['playlist'] = df['playlist'].str.replace("  ", " ")

df.head()
```
